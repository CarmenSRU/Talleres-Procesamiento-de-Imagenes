````python
import cv2
import numpy as np
import matplotlib.pyplot as plt
````

### 1) Lea la imagen 'MRI.png' descargada anteriormente.

````python
imagen = cv2.imread("MRI.png")
````

### 2) Imprima las filas, columnas y canales de la imagen.
Sugerencia: Use ```np.shape()```

````python
print(f"Filas, columnas y canales: {imagen.shape}")
````

### 3) Realice el cambio de color de BRG a escala de grises.
Lo puede realizar con la función ```cv2.cvtColor()```
e indicando el cambio del espacio de color como ```cv2.COLOR_BGR2GRAY```

````python
gris=cv2.cvtColor(imagen,cv2.COLOR_BGR2GRAY)
````

### 4) Imprima nuevamente las filas, columnas y canales de la imagen.

````python
print(f"Filas, columnas y canales: {gris.shape}")
````

### 5) Grafique la imagen con matplotlib
Nota: para evitar el falso color por defecto de matplotlib y el ajuste automático de escalas de grises utilice los siguiente argumentos: 

```plt.imshow(ima, cmap='gray', vmin=0, vmax=255)```

donde cmap='gray' indica el mapa de color en escala de grises, vmin el valor mínimo de nivel de gris y vmax el valor máximo.



````python
plt.imshow(gris, cmap='gray', vmin=0, vmax=255)
plt.show()
````

### 6) Obtenga el histograma de la imagen y grafíquelo
Nota: Como la imagen en escala de grises es de un solo canal solo se debe realizar para el canal 0

````python
histograma = cv2.calcHist([gris],[0],None,[256],[0,256])
plt.figure()
plt.plot(histograma)
plt.grid()
plt.show()
````

### 7) Realice la ecualización de la imagen mediante la ecualización del histograma

````python
equ = cv2.equalizeHist(gris)
````

### 8) Grafique la imagen ecualizada con las mismas consideraciones del punto 5.

````python
plt.imshow(equ, cmap='gray', vmin=0, vmax=255)
plt.show()
````

### 9) Obtenga el histograma de la imagen ecualizada y grafíquelo

````python
histecu = cv2.calcHist([equ],[0],None,[256],[0,256])
plt.figure()
plt.plot(histecu)
plt.grid()
plt.show()
````

### 10) Guarde el resultado obtenido como una imagen con extensión .png

````python
cv2.imwrite('MRIEcualizada.png',equ)
````
